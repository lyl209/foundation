```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def sign(x):
    if x > 0:
        return 1
    else:
        return -1
```

```python
def getx():
    x = np.random.uniform(-1,1)
    return x
```

```python
def gety(x):
    flip = random.choice([-1, 1, 1, 1, 1])
    #print flip
    if x > 0:
        return 1*flip
    else:
        return -1*flip
```

```python
import math

def Eout(s, t):
    return 0.5+0.3*s*(math.fabs(t)-1)
```

```python
def Ein(test_data, s, t):
    count = 0
    for d in test_data:
        h = s*sign(d - t)
        y = gety(d)
        #print 'h, y', h, y
        if h != y:
            count = count + 1
    #print count
    return count*1.0/len(test_data)
    
```

```python
def gen_theta(data):
    t = []
    #t.append(float("inf"))
    t.append(-1)
    for i in xrange(len(data)-1):
        t.append((data[i]+data[i+1])/2.0)
    
    #t.append(-1*float("inf"))
    t.append(1)
    return t
```

```python
data = [0.1,0.2]
gen_theta(data)
```

```python
def gen_data(n):
    data = []
    for i in xrange(n):
        data.append(getx())
    data.sort()
    return data
```

```python
data = gen_data(5)
data
```

```python
def find_lowestEin(data):
        
    thresholds = gen_theta(data)
    Ein_list = []
    for t in thresholds:
        for s in [-1,1]:
            #print s, t
            ein = Ein(data, s, t)
            Ein_list.append((s, t, ein))
    
    #print Ein_list
    
    # Find lowest Ein s and t
    lowest = min(Ein_list, key = lambda t: t[2])
    return lowest
```

```python
data = [-0.4, -0.3, 0.2, 0.4, 0.6]
Ein(data, -1, 0.1)

print '(s, t, Ein) =', find_lowestEin(data)
```
